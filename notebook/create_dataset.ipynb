```python
from PIL import Image, ImageDraw
import random
import os
from math import cos, sin

# 画像サイズ
WIDTH = 600
HEIGHT = 600

# 画像保存用のディレクトリ作成
if not os.path.exists("../VisDiffBench-release/mine/pentagons"):
    os.makedirs("../VisDiffBench-release/mine/pentagons")


def create_pentagon():
    # 新しい画像を作成(白背景)
    image = Image.new("RGB", (WIDTH, HEIGHT), "white")
    draw = ImageDraw.Draw(image)

    # ランダムな色を生成
    color = (random.randint(0, 255), random.randint(0, 255), random.randint(0, 255))

    # ランダムな半径（大きさ）を設定
    radius = random.randint(30, 150)

    # 五角形の中心座標（大きさに応じて配置範囲を調整）
    center_x = random.randint(radius, WIDTH - radius)
    center_y = random.randint(radius, HEIGHT - radius)

    # 五角形の頂点を計算
    angle = random.uniform(0, 2 * 3.14159)  # ランダムな回転角度
    points = []

    for i in range(5):
        x = center_x + radius * cos(angle + 2 * 3.14159 * i / 5)
        y = center_y + radius * sin(angle + 2 * 3.14159 * i / 5)
        points.append((x, y))

    # 五角形を描画
    draw.polygon(points, fill=color)

    return image


# 20枚の画像を生成
for i in range(20):
    img = create_pentagon()
    img.save(f"../VisDiffBench-release/mine/pentagons/{i}.png")
    print(f"Generated pentagon {i}.png")
```

```python
from PIL import Image, ImageDraw
import random
import os
from math import cos, sin

# 画像サイズ
WIDTH = 600
HEIGHT = 600

# 画像保存用のディレクトリ作成
if not os.path.exists("../VisDiffBench-release/mine/hexagons"):
    os.makedirs("../VisDiffBench-release/mine/hexagons")


def create_hexagon():
    # 新しい画像を作成(白背景)
    image = Image.new("RGB", (WIDTH, HEIGHT), "white")
    draw = ImageDraw.Draw(image)

    # ランダムな色を生成
    color = (random.randint(0, 255), random.randint(0, 255), random.randint(0, 255))

    # ランダムな半径(大きさ)を設定
    radius = random.randint(30, 150)

    # 六角形の中心座標(大きさに応じて配置範囲を調整)
    center_x = random.randint(radius, WIDTH - radius)
    center_y = random.randint(radius, HEIGHT - radius)

    # 六角形の頂点を計算
    angle = random.uniform(0, 2 * 3.14159)  # ランダムな回転角度
    points = []

    for i in range(6):  # 6頂点に変更
        x = center_x + radius * cos(angle + 2 * 3.14159 * i / 6)  # 6で割る
        y = center_y + radius * sin(angle + 2 * 3.14159 * i / 6)  # 6で割る
        points.append((x, y))

    # 六角形を描画
    draw.polygon(points, fill=color)

    return image


# 20枚の画像を生成
for i in range(20):
    img = create_hexagon()
    img.save(f"../VisDiffBench-release/mine/hexagons/{i}.png")
    print(f"Generated hexagon {i}.png")
```

```python
from PIL import Image, ImageDraw
import glob
import matplotlib.pyplot as plt


def create_image_grid(image_paths, num_cols=5):
    # 画像を読み込む
    images = [Image.open(path) for path in sorted(image_paths)]

    # 最初の画像のサイズを取得
    w, h = images[0].size

    # 枠線の太さ
    border = 2

    # グリッド画像を作成（枠線用に少し大きめに）
    grid = Image.new(
        "RGB", (w * num_cols + border * (num_cols + 1), h + border * 2), "black"
    )

    # 画像を配置
    for idx, img in enumerate(images[:num_cols]):
        # 枠線の分だけずらして配置
        x = idx * w + border * (idx + 1)
        y = border
        grid.paste(img, (x, y))

    return grid


# 画像のパスを取得
pentagon_paths = sorted(glob.glob("../VisDiffBench-release/mine/pentagons/*.png"))[:5]
hexagon_paths = sorted(glob.glob("../VisDiffBench-release/mine/hexagons/*.png"))[:5]

# グリッドを作成
pentagon_grid = create_image_grid(pentagon_paths)
hexagon_grid = create_image_grid(hexagon_paths)

# 表示
plt.figure(figsize=(20, 8))

plt.subplot(2, 1, 1)
plt.imshow(pentagon_grid)
plt.axis("off")
plt.title("Pentagons")

plt.subplot(2, 1, 2)
plt.imshow(hexagon_grid)
plt.axis("off")
plt.title("Hexagons")

plt.tight_layout()
plt.show()
```
