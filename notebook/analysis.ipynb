```python
import json
import matplotlib.pyplot as plt
import seaborn as sns
import statistics

dataset_name = "easy"
plot_metrics = "accuracy"

with open(f"../results_{dataset_name}.json", "r", encoding="utf-8") as f:
    data = json.load(f)

metrics = data["metrics"]

accuracies = [m["accuracy"] for m in metrics]
acc_top_1 = [m["accuracy_top_1"] for m in metrics]
acc_top_5 = [m["accuracy_top_5"] for m in metrics]

metrics_of_interest = {
    "accuracy": accuracies,
    "accuracy_top_1": acc_top_1,
    "accuracy_top_5": acc_top_5,
}

print("mean and std of metrics:")
for metric, values in metrics_of_interest.items():
    print(f"{metric}: {statistics.mean(values):.2f} ± {statistics.stdev(values):.2f}")

plt.figure(figsize=(10, 5))
sns.histplot(metrics_of_interest[plot_metrics], bins=10)
plt.title(f"{plot_metrics.title()} distribution (Dataset: {dataset_name})")
plt.xlabel(plot_metrics.title())
plt.ylabel("Count")
plt.show()
```
